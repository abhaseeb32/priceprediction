```python
import numpy as np
import pandas as pd
```

```python
home_prices = pd.read_csv(r'E:\Courses\Kaggle Datasets\train.csv')
```

### Checking for Null values now:

```python
home_prices.isnull().values.any()
```

```python
home_prices.isnull().sum().sum()
```

### Dropping Null values:

```python
home_price = home_prices.dropna(axis=1)
```

```python
home_price.head()
```

```python
home_price.isnull().sum().sum()
```

```python
home_price.describe()
```

### find prediction target

```python
#here in this data its "SalePrice" So,

y= home_price.SalePrice
print(y)
```

### Finding the predictive features here for 'y'

```python
#we will find those features and save them in 'features' variable,

features = ["LotArea", "YearBuilt", "1stFlrSF", "2ndFlrSF","FullBath", "BedroomAbvGr", "TotRmsAbvGrd"]

#saving them in a value:

X = home_price[features]
print(X)
```

```python
X.head()
```

### Now we will create a model for predictions:

```python
#specifying the model:

from sklearn.tree import DecisionTreeRegressor
price_model = DecisionTreeRegressor(random_state=1)

# Fit the model
price_model.fit(X,y)
```

### predictions:

```python
predictions = price_model.predict(X)
print(predictions)
```
